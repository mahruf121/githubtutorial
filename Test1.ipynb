```python
import pandas as pd
```

```python
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df[df['location']=='Pakistan'].groupby(['date'])[['new_tests', 'new_cases']].max().plot(figsize=(15,4))
```

```python
bd = df[df['location']=='Bangladesh']
bd.pivot
```

```python
bp = df[df['location'].isin(['Bangladesh', 'Pakistan'])]
```

```python
bp.pivot(index='date', columns='location', values='new_cases_per_million').plot(figsize=(15,4))
```
